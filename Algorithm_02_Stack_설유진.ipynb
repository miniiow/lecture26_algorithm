### Stack의 연산 구현

```python
# 필요한 데이터 정의
statck = []

while True:
    # 메뉴를 출력
    print('============================================')
    print('1.Push       2.Pop      3.Peek      0.Exit')
    print('============================================')
    
    # 입력을 받는다
    menu = int(input('>> 연산 메뉴 선택 : '))
    # print(menu)
    
    # 입력에 따라 stack 연산을 수행하고 스택의 상태를 출력한다
    if menu == 0:  # Exit
        # print('정수형 스택 종료')
        break
    elif menu == 1:  # Push
        data = int(input('>> [Push] 추가할 데이터 입력 : '))
        statck.append(data)
    elif menu == 2:  # Pop
        data = statck.pop()
        print('[Pop] 가져온 데이터 : ', data)
    elif menu == 3: # Peek
        data = statck[len(statck)-1]
        print('>> top 데이터 확인 : ',data)
        continue  # 여기에서 더 진행하지 말고 반복문으로 돌아가
    
    print('[Stack]', statck)
    
print('정수형 스택 종료')
# 위의 과정을 0. Exit을 선택할때까지 반복한다
```
