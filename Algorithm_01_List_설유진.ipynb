### 파이썬 알고리즘 List 연습

```python
# 파이썬 List 연습 : 항목 접근

test_list = [1,2,3,'one','two','three']
print(test_list[0])
print(test_list[3])
# 파이썬은 인덱스값 음수 사용 가능 >> 음수면 뒤에서부터 시작
print(test_list[-1])
print(test_list[-3])

print('==========')
for item in test_list:
    print(item)
```

```python
# 파이썬 List 연습 : 항목 삽인과 삭제
# insert(index, item) : index 위치에 항목을 추가
# append() : 리스트의 가장 뒤에 항목을 추가
# remove(item) : 리스트에서 첫 번째로 나오는 item 삭제
# pop(index) : index 위치의 항목 삭제
# pop() : 리스트의 마지막 항목 삭제

test_list = [1,2,3,'one','two','three']
test_list.insert(3,4)
print(test_list)
test_list.append('four')
print(test_list)
test_list.remove('one')
print(test_list)
# pop을 사용할때는 데이터를 꺼내서 다른곳에 사용할 수도 있으므로 변수에 할당하는 것이 좋다
item = test_list.pop(0)
print(item)
print(test_list)
item = test_list.pop()
print(test_list)
```

### 2026.03.23 숙제 : 리스트와 반복구조 연습

```python
# 리스트와 반복 구조
# 주어진 데이터가 어디에 저장되어 있는지 출력
data_list = [23,45,21,14,2,67]
target = 14

index = 0
while index <= 5:
    if data_list[index] == target:
        break
    # index 값은 0부터 1씩 더해진다
    index += 1
print(index)

while index <= 5 and data_list[index] != target:
    index += 1
print(index)

# if index > 5 :
#     print('값을 찾지 못했습니다')
# else :
```

```python
# 리스트에 저장된 숫자 중 가장 큰 숫자 출력 (최대값)

# 인덱스 값으로 접근
data_list = [23,45,21,14,2,67]
max = data_list[0]
max_index = 0
for i in range(1, 6):
    if data_list[i] > max:
        max = data_list[i]
        max_index = i
print(max)
# 최대값의 인덱스 번호 알고싶다면
# print(max_index)


# data_list 자체가 시퀀스로 지정
for data in data_list:
    if data > max:
        max = data
print(max)
```

```python
# 리스트에 저장된 숫자 중 가장 작은 숫자 출력(최솟값)

# 인덱스 값으로 접근
data_list = [23,45,21,14,2,67]
min = data_list[0]
for i in range(1, 6):
    if min > data_list[i]:
        min = data_list[i]
print(min)


# data_list 자체가 시퀀스로 지정
for data in data_list:
    if data < min:
        min = data
print(min)
```

```python
# 리스트에 저장된 숫자 중 입력된 숫자보다 큰 수가 몇 개 있는지 구하기

data_list = [23,45,21,14,2,67]
num = int(input('숫자를 입력하세요:'))
count = 0

for i in range(0, 6):
    if num < data_list[i]:
        # print(data_list[i])
        count += 1
print(f'입력한 숫자보다 큰 수는 총 {count}개 입니다.')
```

```python
# 리스트에 저장된 모든 값 더하여 출력(합)

data_list = [23,45,21,14,2,67]
sum = 0
for i in range(0, 6):
    print(i)
    sum += data_list[i]
print(sum)
```

```python
# 리스트에 저장된 숫자를 역순으로 저장하여 출력
# 리스트 새로 만들지 말고 자기 리스트에서 출력해야 함

data_list = [23,45,21,14,2,67]
index = 5
temp = 0
for i in range(0, 3, 1):
    # print(i)
    # 리스트 마지막 값
    # temp = data_list[index]
    # data_list[index] = data_list[i]
    # data_list[i] = temp
    # 한줄로 압축가능함
    data_list[i], data_list[index] = data_list[i], data_list[index]
    index -= 1
print(data_list)
```
