### **탐색**

*선형자료구조일때 두가지 방법으로 탐색할 수 있다*  <br>
> #### **순차탐색**
> *처음부터 하나하나 해당하는 값인지 확인한다*
> #### **이진탐색**
> *정렬이 되어있는 자료구조를 반씩 쪼개면서 해당하는 값이 클때는 오른쪽의 데이터를, 작을때는 왼쪽의 데이터로 탐색함*

```python
data = [2,6,11,13,18,20,22,27,29,30,34,38,41,42,45,47]
```

```python
# 이진탐색_Binary Search_recursive

def binary_search_recursive(A,key,low,high):
    if (low > high):  # low값과 high값이 뒤집힌 경우 = 값을 찾지 못한경우
        return -1     # 찾는값이 없어 -1을 반환한다

    middle = (low + high) // 2

    if (key == A[middle]):
        return middle
    elif (key < A[middle]):
        return binary_search_recursive(A,key,low,middle-1) # high값을 middle의 -1번째로 변경하여 다시 함수를 불러온다
                                                           # = 왼쪽값을 탐색한다
    else:
        return binary_search_recursive(A,key,middle+1,high) # low값을 middle의 +1번째로 변경하여 다시 함수를 불러온다
                                                           # = 오른쪽값을 탐색한다
```

```python
print(binary_search_recursive(data,27,0,15))
```

```python
# 이진탐색_Binary Search_while loof

def binary_serch(A, key):  # A:데이터 / key:찾을 값
    # low와 high의 초기값 설정
    low = 0           # 데이터의 처음 = 가장 작은 숫자
    high = len(A)-1   # 데이터의 마지막 = 가장 큰 숫자
    while low <= high: # low가 high보다 작거나 같다는 것은 아직 볼 데이터가 남아있다는 상태라는 뜻이기때문에 계속 반복한다
                       # = low와 high가 뒤집히기 전까지는 계속 반복해라
        middle = (low + high) // 2    # 중앙값 찾기:2로 나눴을때 몫 찾기 ex) 0+15//2 = 7반환
        if A[middle] == key:
            return middle     # 데이터의 middle번째 값이 key와 같다면 원하는 값을 찾았다는 뜻이니 middle을 반환한다
        elif A[middle] > key: # middle값이 key보다 작음:왼쪽 부분을 탐색해야 함
            high = middle-1  # high를 middle의 -1로 변경하여 왼쪽으로 탐색하게 바꾼다
        else:
            low = middle+1   # middle값이 key보다 클경우:오른쪽탐색을 해야하기때문에 low를 middle+1한다
    return -1  # 찾는값이 없을경우 -1을 반환한다
```

```python
print(binary_serch(data,27))
```

```python
# 개발팀에서 테스트:스모크테스트

num = int(input('찾을 숫자: '))
index = binary_search_recursive(data, num, 0,len(data))

if index == -1:
    print('없는 숫자 입니다.')
else:
    print(f">>> 숫자 {num}은 index {index}에 있습니다.")

index = binary_serch(data, num)
if index == -1:
    print('없는 숫자 입니다.')
else:
    print(f">>> 숫자 {num}은 index {index}에 있습니다.")
```
