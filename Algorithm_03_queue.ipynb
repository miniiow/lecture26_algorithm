```python
# 정수형 큐 연산 파이썬 구현 실습

from collections import deque
queue = deque()

def enqueue(data):
    queue.append(data)

def dequeue():
    data = queue.popleft()
    if queue:
        return data
    else:
        return -1

while True:
    print('===========================================')
    print(' 1.Enqueue  2.Dequeue  3.Get Front  0.Exit')
    print('===========================================')
    menu = int(input('>> 연산 메뉴 선택 : '))
    if menu == 0:
        break
        
    if menu == 1:   # Enqueue
        data = int(input('>> 추가할 데이터 입력 : '))
        enqueue(data)
        
    if menu == 2:  # Dequeue
        # print('가져온 데이터 :', dequeu() if queue else 'Queue가 비어 있습니다.')
        if queue:
            dequeue()
            print('가져온 데이터 : ', data)
        else:
            print('Queue가 비어 있습니다.')

    if menu == 3:   # Get Front
        print(queue[0] if queue else 'Queue가 비어 있습니다.')
        # if queue:
        #     print(queue[0])
        # else:
        #     print('Quere가 비어 있습니다.')

    print(queue)
print('정수형 Queue 종료')
```
