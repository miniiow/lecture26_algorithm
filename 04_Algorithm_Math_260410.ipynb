### **수학(Math)**

```python
# 소수구하는 함수 만들기 1

def is_prime_number(number):
    num_is_prime = True
    if number == 1:
        num_is_prime = False
    else:
        for i in range(2,number):
            if number% i == 0:
                num_is_prime = False
                break
    return num_is_prime
```

```python
for i in range(10):
    print(i, is_prime_number(i))
```

```python
# 소수구하는 함수 만들기 2

def is_prime_number2(number):
    if number == 1:
        return False
    else:
        for i in range(2,number):
            if number% i == 0:
                return False
                break
    return True
```

```python
for i in range(10):
    print(i, is_prime_number2(i))
```

```python
# 소수구하는 함수 개선 1
# 2의 배수를 소수가 아니다

def is_prime_number3(number):
    if number == 2: return True
    if number == 1 or number%2 == 0: return False

    for i in range(3, number, 2):
        if number% i == 0:
            return False
            break
    return True
```

```python
for i in range(10):
    print(i, is_prime_number3(i))
```

### **진법**

```python
# 정수를 입력받아 특정 진법의 수로 변환하는 알고리즘 구현

# 정수 N을 받아서 base에 대한 진법의 수로 변환
def convert_base(N, base):
    # N을 base로 나눈 몫이 0이 될때까지 나누기
    # 나머지를 진법에 해당하는 숫자로 변환하여 리스트에 저장
    # N : 몫을 저장
    # new_list : 나머지를 저장하는 리스트
    new_number = ''
    
    # 사용할 숫자와 알파벳 정의(36진법까지 가능) : 자료구조를 만들어서 하나하나 입력하지 않아도 됨
    digits = '0123456789ABCDEFGHIGKLMNOPQRSTUVWXYZ'
    while N != 0:
        # new_list.append(digits[N % base])
        # 따로 뒤집지 말고 그냥 새로운 값 앞에 저장 (파이썬에서 가능함)
        new_number = digits[N % base] + new_number
        N //= base
    return new_number

print(convert_base(25,2))
print(convert_base(25,8))
print(convert_base(25,10))
print(convert_base(25,16))
```
