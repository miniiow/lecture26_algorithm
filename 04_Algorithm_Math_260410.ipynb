### **수학(Math)**

```python
# 소수구하는 함수 만들기 1

def is_prime_number(number):
    num_is_prime = True
    if number == 1:
        num_is_prime = False
    else:
        for i in range(2,number):
            if number% i == 0:
                num_is_prime = False
                break
    return num_is_prime
```

```python
for i in range(10):
    print(i, is_prime_number(i))
```

```python
# 소수구하는 함수 만들기 2

def is_prime_number2(number):
    if number == 1:
        return False
    else:
        for i in range(2,number):
            if number% i == 0:
                return False
                break
    return True
```

```python
for i in range(10):
    print(i, is_prime_number2(i))
```

```python
# 소수구하는 함수 개선 1
# 2의 배수를 소수가 아니다

def is_prime_number3(number):
    if number == 2: return True
    if number == 1 or number%2 == 0: return False

    for i in range(3, number, 2):
        if number% i == 0:
            return False
            break
    return True
```

```python
for i in range(10):
    print(i, is_prime_number3(i))
```
