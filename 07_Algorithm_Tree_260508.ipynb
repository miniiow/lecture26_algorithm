> **1. 이진트리 (Binary Tree)**  
> *하나의 노드가 서브트리를 최대 2개가지만 가지는 트리*  <br>
>    + **포화이진트리(Full binary tree)**  
>        *트리의 각 레벨에 노드가 꽉 차있는 이진 트리*<br><br>  
>    + **완전이진트리 (complete binary tree)**  
>        *높이가 h일때 레벨 1부터  h-1까지는 노드가 모두 채워짐*

```python
class TreeNode:
    def __init__(self, data,left, right):
        self.data = data
        self.left = left
        self.right = right
    def set_left(self, left):
        self.left = left
    def set_right(self, right):
        self.right = right
```

```python
my_tree = TreeNode('A', None, None)
new_node = TreeNode('B', None, None)
my_tree.set_left(new_node)
new_node = TreeNode('C', None, None)
my_tree.set_right(new_node)

cur_node = my_tree.left # B값을 가지고 있는 변수명이 없는데 my_tree left에 B주소가 있으니 그거 사용함
new_node = TreeNode('D', None, None)
cur_node.set_left(new_node)
cur_node.set_right(TreeNode('E', None, None))
```

```python
d = TreeNode('D', None, None)
e = TreeNode('E', None, None)
b = TreeNode('B', d, e)

f = TreeNode('F', None, None)
c = TreeNode('C', f, None)

a = TreeNode('A', b, c)

my_binary_tree = a
```

```python
# preorder (선위순회)

def preorder(node):
    print(node.data, end=' ')
    # if node.left != None: preorder(node.left)
    # if node.right != None: preorder(node.right)
    if node.left: preorder(node.left)
    if node.right: preorder(node.right)

preorder(my_binary_tree)
```

```python
# inorder (중위순회)

def inorder(node):
    if node == None: return

    inorder(node.left)
    print(node.data, end=' ')
    inorder(node.right)

inorder(my_binary_tree)
```

```python
# postorder (후위순회)

def postorder(node):
    if node == None : return
    
    postorder(node.left)
    postorder(node.right)
    print(node.data, end=' ')

postorder(my_binary_tree)
```

```python
tree = {}
tree['A'] = ['B', 'C']
tree['B'] = ['D', 'E']
tree['C'] = ['F', None]
tree['D'] = [None, None]
tree['E'] = [None, None]
tree['F'] = [None, None]

my_tree = 'A'
```

```python
# preorder (전위순회)

def preorder(node):
    if node is None: return

    print(node, end=' ')

    preorder(tree[node][0])
    preorder(tree[node][1])

preorder(my_tree)
```

```python
# inorder (중위순회)

def inorder(node):
    if node == None: return

    inorder(tree[node][0])
    print(node, end=' ')
    inorder(tree[node][1])

inorder(my_tree)
```

```python
# postorder (후위순회)

def postorder(node):
    if node == None : return
    
    postorder(tree[node][0])
    postorder(tree[node][1])
    print(node, end=' ')

postorder(my_tree)
```

### Binary Search Tree

```python
class STreeNode:
    def __init__(self, key,left, right):
        self.key = key
        self.left = left
        self.right = right
```

```python
# BST 생성

t3 = STreeNode(3, None, None)
t12 = STreeNode(12, None, None)
t7 = STreeNode(7, t3, t12)
t27 = STreeNode(27, None, None)
t31 = STreeNode(31, t27, None)
t26 = STreeNode(26, None, t31)
t18 = STreeNode(18, t7, t26)

my_bst = t18
```

```python
def search_bst(n, key):
    if n == None : return None
    if key == n.key: return n
    if key < n.key:
        return search_bst(n.left, key)
    else:
        return search_bst(n.right, key)
    
result = search_bst(my_bst, 40)
if result: print(result.key)
else:
    print('없는 key입니다.')
```

```python
def search_bst_iter(n, key):
    while n != None:
        if key == n.key: return n
        elif key < n.key:
            n = n.left
        else:
            n = n.right
    return None

result = search_bst_iter(my_bst, 12)
if result: print(result.key)
else:
    print('없는 key입니다.')
```

```python
# 숫자를 순서대로 입력받아 binary search tree를 구성하기: insert_bst()함수 활용

def insert_bst(n, key):
    while n != None:
        parent = n
        if key == n.key:
            return False
        elif key < n.key:
            n = n.left
        else:
            n = n.right
    return parent

# preorder (선위순회)
def preorder(node):
    print(node.key, end=' ')
    if node.left: preorder(node.left)
    if node.right: preorder(node.right)

turn = int(input('몇 번 입력? : '))
for i in range(turn):
    input_num = int(input('숫자를 입력하세요 : '))
    result = insert_bst(my_bst, input_num)

    if result is False:
        print('이미 있는 값입니다.')
    elif input_num < result.key:
        result.left = STreeNode(input_num, None, None)
        print(f'{result.key}의 왼쪽에 {input_num}입력되었습니다.')
    else:
        result.right = STreeNode(input_num, None, None)
        print(f'{result.key}의 오른쪽에 {input_num}입력되었습니다.')
    preorder(my_bst)
    print()
    
print('[종료]')
```
