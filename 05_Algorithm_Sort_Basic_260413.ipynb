### **선택 정렬(Selection Sort)**
*정렬 안 된 부분에서 최소(최대)값을 선택 후 정렬 안된 부분의 처음 데이터와 교환*

```python
# selectioin sort
# 정렬이 안된 부분 (start 이후)에서 가장 작은 수를 찾아서
# start에 있는 수와 swap
# start를 += 1

def print_step(start, A):
    print(f'{start:2} 단계: {A}')

# 리스트에 있는 가장 작은 값 찾아내기
def select(A, start):
    min = A[start]
    min_index = start

    for i in range(start+1, len(A)):
        if A[i] < min:
            min = A[i]
            min_index = i
    return min_index

def selection_sort(data_list):
    for start in range(len(data_list)-1):
        min_index = select(data_list, start)
        data_list[start], data_list[min_index] = data_list[min_index], data_list[start]
        print_step(start, data_list)


A = [23,5,44,1,6]
print('원래 데이터: ', A)
selection_sort(A)
print('정렬 후 데이터: ', A)
```

### **삽입 정렬(Insertion Sort)**
*정렬 안 된 부분 원소를 하나씩 정렬된 부분에 적절한 위치에 '삽입'하여 정렬*

```python
# insertion sort

def insertion_sort(A):
    for cur_index in range(1, len(A)):
        # 해당 데이터의 위치를 찾아서 insertion 진행
        key = A[cur_index]
        i = cur_index - 1
        while i >= 0 and A[i] > key:
            A[i+1] = A[i]
            i -= 1
        # 작거나 같은 데이터의 오른쪽에 저장
        A[i+1] = key
        print_step(cur_index, A)
```

```python
A = [23,5,44,1,6]
print('원래 데이터: ', A)
insertion_sort(A)
print('정렬 후 데이터: ', A)
```

### **버블 정렬(Bubble Sort)**
*정렬이 안된 부분의 두 개의 인접한 데이터 a, b가 a<b가 되도록 Swap 연산 수행*

```python
# Bubble Sort

def bubble_sort(A):
    for last in range(len(A)-1, 0, -1):
        is_change = False
        for i in range(last):
            if A[i] > A[i+1]:
                A[i], A[i+1] = A[i+1], A[i]
                is_change = True
        
        print_step(last, A)
        if not is_change: return
```

```python
A = [1,23,5,44,1,6]
print('원래 데이터: ', A)
bubble_sort(A)
print('정렬 후 데이터: ', A)
```
